```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['species']=iris.target
df.head()
```

```python
x=df.drop('species',axis=1)
x.head()
```

```python
from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=3,random_state=42)
kmeans.fit(x)
kmeans_labels=kmeans.labels_
kmeans_labels
```

```python
plt.scatter(x["sepal length (cm)"],x["petal length (cm)"],c=kmeans_labels,cmap='viridis')
plt.xlabel("sepal length")
plt.ylabel("petal length")
plt.title("kMeans Clustering")
plt.show()
```



*   K-Means is an unsupervised machine learning algorithm used to group similar data points into a specified number of clusters (K).
*   The Iris dataset contains numerical features such as sepal length, sepal width, petal length, and petal width, which are suitable for distance-based clustering.
The dataset naturally contains three groups of flowers, so choosing K = 3 allows the algorithm to form three clusters.



```python
from sklearn.cluster import AgglomerativeClustering

hierarchical=AgglomerativeClustering(n_clusters=3)
hierarchical.fit(x)
hierarchical_labels = hierarchical.labels_
hierarchical_labels
```

```python
plt.scatter(x["sepal length (cm)"],x["petal length (cm)"],c=hierarchical_labels,cmap="viridis")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Hierarchical Clustering")
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

linked = linkage(x, method='ward')
dendrogram(linked)
plt.title("Dendrogram for Iris Dataset")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```



*   Hierarchical Clustering is an unsupervised learning algorithm that groups similar data points by merging the closest clusters step by step. The clustering process is represented using a dendrogram
*   The Iris dataset is small, containing only 150 flower samples, making it efficient for hierarchical clustering.
It works well with the numerical features of the Iris dataset, such as sepal length, sepal width, petal length, and petal width.It groups similar flowers based on their feature values without using the species labels
